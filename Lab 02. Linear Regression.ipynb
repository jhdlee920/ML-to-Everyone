# 1. Build graph using TF operations

## H(x) = Wx+b

```python
import tensorflow as tf
```

```python
# X and Y data
x_train= [1,2,3]
y_train= [1,2,3]
#학습할 데이터가 주어졌다!

# Variable이란 node로 정의
# 값을 모르기 때문에 random_normal인 shape과 [1]을 준다.
W=tf.Variable(tf.random_normal([1]),name='weight')
b=tf.Variable(tf.random_normal([1]),name='bias')
# Our hypothesis Wx+b
hypothesis = x_train * W + b
```

```python
# cost function
# reducce_mean은 평균을 구하는 node
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
```

## Gradient Descent

```python
# Minimize하는 방법 중 하나
```

```python
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost)
```

# 2. Run/update graph and get results

```python
# Launch the graph in a session
sess=tf.Session()
```

```python
# variable 을 사용하기 이전에는 항상 initialize 해줘야함
# Initializes global variables in the graph
sess.run(tf.global_variables_initializer())
```

```python
# Fit the line
for step in range(2001): # 2001까지 스텝을 도는데
    sess.run(train)
    if step % 20 ==0:    # 20번에 한 번씩 print를 한다.
        print(step,sess.run(cost),sess.run(W),sess.run(b))
```

# Placeholder를 활용한 Linear Regression

```python
X=tf.placeholder(tf.float32)
Y=tf.placeholder(tf.float32)
```

```python
for step in range(2001):
    cost_val, W_val, b_val, _ = \ # train에 나오는 value는 중요하지 않으므로 표시x(언더바)
        sess.run([cost,W,b,train],feed_dict={X:[1,2,3],Y:[1,2,3]})
    if step % 20 ==0:
        print(step,cost_val,W_val,b_val)
```
